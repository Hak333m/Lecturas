Load the necessary packages

```
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
% matplotlib inline
```

Read the file and store it in a variable.

```
with open("README.org") as f:
    content = f.readlines()
```

```
content
```

```
cleanedLines = ''.join([' '.join([x.replace('#','').replace('.','') for x in line.replace('\n','').split() if ('#' in x and '+' not in x and '[' not in x )])   for line in content])
```

```
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(cleanedLines)
```

```
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

We can also search per month

```
def wordCloudMonth(month,lines):
  monthlines = []
  add = False
  for line in lines:
    if '** ' +  month in line:
      print(line)
      add = True
    elif add==True and '**' in line:
      add = False
    elif add==True:
      monthlines.append(line)
  cleanedLines = ' '.join([' '.join([x.replace('#','').replace('.','') for x in line.replace('\n','').split() if ('#' in x and '+' not in x and '[' not in x )])   for line in monthlines])
  wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(cleanedLines)
  plt.figure()
  plt.imshow(wordcloud, interpolation='bilinear')
  plt.axis("off")
  plt.show()
```

```
month = 'Noviembre 2019'
```

```
wordCloudMonth(month,content)
```
